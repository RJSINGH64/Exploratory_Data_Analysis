```python
#segregate the cate and num data
num_values=[i for i in df.columns if df[i].dtype !="O"]
cate_values=[j for j in df.columns if df[j].dtype =="O"]
```

```python
df[cate_values]
```

```python
num_values
```

```python
#check the distribution of the numeric dataset
num_data=df[num_values]
```

```python
for i in num_values:
    sns.kdeplot(num_data[i])
    plt.xlabel(i)
```

```python
plt.figure(figsize=(15,15))
plt.suptitle("Univariate analysis of numericle feature",fontsize=20 , fontweight="bold" ,alpha=0.8 , y=1)
for i in range(0,len(num_values)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df[num_values[i]],shade=True , color="r")
    plt.xlabel(num_values[i])
    plt.tight_layout()
```

```python
for i in num_values:
    sns.kdeplot(num_data[i])
    plt.xlabel(i)
    plt.ylabel("Count")
    plt.title("Numerical data")
    plt.show()
```
